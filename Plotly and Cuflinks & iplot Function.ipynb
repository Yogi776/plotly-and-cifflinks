# Why we used Plotly Library?

#### Plotly is a library that allows you to create interactive plots that can be use in dashboards or websites

# Import library

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from plotly.offline import iplot
import plotly as py
import plotly.tools as tls
import cufflinks as cf
```

```python
df=pd.DataFrame(np.random.randn(100,3),columns=['X','Y','Z'])
df.head(5)
```

```python
df['X']=df['X'].cumsum() + 20
df['Y']=df['Y'].cumsum() + 20
df['Z']=df['Z'].cumsum() + 20
```

```python
df.head()
```

```python
py.offline.init_notebook_mode(connected=True)
```

```python
cf.go_offline()
```

```python
df.iplot()
```

# Mode Function
